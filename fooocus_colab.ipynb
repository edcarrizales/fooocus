```python
# Solo necesitas instalar gdown una vez
!pip install -q gdown ipywidgets

import gdown
import os
import ipywidgets as widgets
from IPython.display import display

# Diccionario con los checkpoints disponibles
checkpoints = {
    "cyberprimeXL_v10": "1DqiKaHGXfn5gKU3Qg49GLdHny6yDT1bQ",
    "unstableIllusion_sdxxxl": "1mHyUdM0Riqo7uMiLs5zWdvQqUrkH_ysy",
    "sdxxxl_v30": "1VzuDW0vxjr-C20oF-Q8SmTLLZI26fbgv",
    "fabledIllusionXXXLSFWNSFWPorn_v3PhotoNoir": "1qE2tufq6-LqgPrIzXLcrfysKazjbsfJE"
}

# Widget para seleccionar hasta 2 archivos
selector = widgets.SelectMultiple(
    options=checkpoints.keys(),
    description='Selecciona:',
    disabled=False
)

boton_descargar = widgets.Button(description="Descargar Seleccionados")

def on_click(b):
    seleccion = list(selector.value)
    if len(seleccion) > 2:
        print("⚠️ Solo puedes seleccionar hasta 2 modelos.")
        return

    os.makedirs("/content/checkpoints", exist_ok=True)

    for nombre in seleccion:
        file_id = checkpoints[nombre]
        destino = f"/content/checkpoints/{nombre}.safetensors"
        print(f"📥 Descargando {nombre}...")
        gdown.download(id=file_id, output=destino, quiet=False)

    if seleccion:
        print("✅ Descargas completadas.")

boton_descargar.on_click(on_click)

# Mostrar el widget
display(selector, boton_descargar)
```
